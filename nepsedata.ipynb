```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('nepse.csv')
df
```

```python
print("Number of Rows: ",df.shape[0])
print("Number of Columns: ",df.shape[1])
```

```python
df.info()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data = df)
```

```python
df.isnull().sum()
```

```python
#Line plot of Closing Prices over Time


# Converting Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Plot Closing Prices over Time
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], color='blue')
plt.title('Closing Prices over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()
```

```python
#Histogram of Daily Trading Volume

# Converting Volume column to numeric format
df['Volume'] = pd.to_numeric(df['Volume'].str.replace(',', ''))

# Histogram of Daily Trading Volume
plt.figure(figsize=(10, 6))
plt.hist(df['Volume'], bins=30, color='green', edgecolor='black')
plt.title('Histogram of Daily Trading Volume')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Bar plot of Daily Percent Changes

# Replace '-- ' with '0.00' and convert to numeric format
df['Percent Change'] = pd.to_numeric(df['Percent Change'].str.replace('-- ', '0.00').str.replace('%', ''))

# Plot Bar plot of Daily Percent Changes
plt.figure(figsize=(10, 6))
plt.bar(df['Date'], df['Percent Change'], color='orange')
plt.title('Daily Percent Changes')
plt.xlabel('Date')
plt.ylabel('Percent Change')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Scatter plot of High vs. Low Prices

# Plot Scatter plot of High vs. Low Prices
plt.figure(figsize=(10, 6))
plt.scatter(df['High'], df['Low'], color='purple')
plt.title('High vs. Low Prices')
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.grid(True)
plt.show()
```

```python
# Box plot of Closing Prices by Month

# Extract month from Date
df['Month'] = df['Date'].dt.month

# Box plot of Closing Prices by Month
plt.figure(figsize=(10, 6))
df.boxplot(column='Close', by='Month')
plt.title('Closing Prices by Month')
plt.xlabel('Month')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()
```

```python
#Density plot of Closing Prices

# Plot Density plot of Closing Prices
plt.figure(figsize=(10, 6))
df['Close'].plot(kind='density', color='brown')
plt.title('Density Plot of Closing Prices')
plt.xlabel('Closing Price')
plt.grid(True)
plt.show()
```

```python
# Calculate the 100-day moving average
df['Close_MA_100'] = df['Close'].rolling(window=100).mean()

# Plot the Close price and its 100-day moving average
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], label='Close Price')
plt.plot(df['Date'], df['Close_MA_100'], label='100-Day Moving Average', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Close Price and 100-Day Moving Average')
plt.legend()
plt.show()
```

```python
# Calculate the 200-day moving average
df['Close_MA_200'] = df['Close'].rolling(window=200).mean()

# Plot the Close price, 100-day moving average, and 200-day moving average
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], label='Close Price')
plt.plot(df['Date'], df['Close_MA_100'], label='100-Day Moving Average', linestyle='--')
plt.plot(df['Date'], df['Close_MA_200'], label='200-Day Moving Average', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Close Price and Moving Averages')
plt.legend()
plt.show()
```
